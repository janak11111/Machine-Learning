```python
# import libraries
import numpy as np
import cv2 as cv
import glob
import random
import matplotlib.pyplot as plt
```

```python
# select from images we have to fetch
images = glob.glob("./data/*.bmp")


# generate random image and apply 6 operation on each 3 images
for i in range(1, 4):
    print("For Image {}".format(i))
    
    # Read image
    img = cv.imread(random.choice(images))
    
    
    # Resize image
    resized_img = cv.resize(img, (1200, 1000), interpolation = cv.INTER_NEAREST)
    plt.imshow(resized_img)
    plt.title("Resized Image")
    plt.show()
    
    
    # Pad image
    padded_img = cv.copyMakeBorder(img, 200, 420, 320, 310, cv.BORDER_CONSTANT)
    plt.imshow(padded_img)
    plt.title("Padded Image")
    plt.show()
    
    
    # Crop image
    cropped_img = img[132:1200, 300:1526]
    plt.imshow(cropped_img)
    plt.title("Cropped Image")
    plt.show()
    
    
    # Gray scale image
    gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    plt.imshow(gray_img)
    plt.title("Gray scale Image")
    plt.show()
    
    
    # Apply contrast 
    contrasted_img = cv.convertScaleAbs(img, alpha=1.8, beta=20)
    plt.imshow(contrasted_img)
    plt.title("Contrasted Image")
    plt.show()
    
    
    # Apply saturation
    saturated_img = cv.convertScaleAbs(img, alpha=1, beta=1)
    plt.title("Saturated Image")
    plt.imshow(saturated_img)
    plt.show()
    print()
    print()
    print()
    
```
