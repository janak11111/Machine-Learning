## Importting Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.arange(13)
Y = (X-6)**2
print(X, Y)
```

### Visulization

```python
plt.style.use('seaborn')
plt.plot(X, Y, label='Line', c='orange')
plt.xlabel('X')
plt.ylabel('F(X)')
plt.legend(fontsize='xx-large')
plt.show()
```

```python
alpha = 0.1
interation = 50
error = []
plt.plot(X, Y)
x = 0


for i in range(interation):
    grad = 2*(x-6)
    x = x-alpha*grad
    y = (x-6)**2
    error.append(y)
    plt.scatter(x, y)
    print(x)
```

```python
plt.plot(error)
```
