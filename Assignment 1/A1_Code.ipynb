```python
# print firdt 5 rows
airfoil.head()
```

```python
# check shape
airfoil.shape
```

```python
# check data types
airfoil.info()
```

```python
# check for duplicate data

airfoil["is_duplicate"]= airfoil.duplicated()
airfoil.is_duplicate.unique()
```

```python
# scale data
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
x = scaler.fit_transform(airfoil.iloc[:, :-1])

airfoil_scaled = pd.DataFrame(x)
airfoil_scaled.columns = airfoil.columns[:-1]


# heatmap for correlation
sns.heatmap(airfoil_scaled.corr(), annot=True)
```

```python
# box plot for each fetures
for i in airfoil_scaled.columns:
  sns.boxplot(airfoil_scaled[i])
  plt.show()
```

```python
## Analysis of Box plots

## Feature Frequency and Suction_side_thickness have more outliers.
## Distribution of chord_length and angle_of_attck are postive skewed.
## Distribution of pressure_level is almost Gaussian.
```

```python
# pair plot
sns.pairplot(airfoil_scaled)
```

```python
## Analysis of Pair plot

## There is no proper correletion between features.
## trends in most cases are  one feature increases and other feature value is constant.
```

```python
# extract target and feture from dataset

features = airfoil_scaled.iloc[:, :-1]
target = airfoil_scaled.iloc[:, -1]
```

```python
# split data into train and test

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# create regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
# MAE loss function from scratch

def MAE(y_actual, y_pred):
  return sum(abs(y_pred - y_actual)) / len(y_pred)

# print loss after function call
print(MAE(y_test, y_pred))
```

```python
# MAE loss using sklearn
print(mean_absolute_error(y_test, y_pred))
```

```python
# compute R2 score of model
print(r2_score(y_test, y_pred))
```

## Preprocessing

```python
# data standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_glass = pd.DataFrame(scaler.fit_transform(glass.iloc[:, :-1]), columns=cols[:-1])
scaled_glass.head()
```

```python
# extract feature and target from dataset

features = scaled_glass
target = glass.iloc[:, -1]
```

```python
# split data into train and test
# set random seeed top 0
from sklearn.model_selection import train_test_split

np.random.seed(0)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

## Create Model  and predict

```python
# make logisitc regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# check accuracy, precision, recall, F1 score

from sklearn.metrics import accuracy_score, precision_score ,recall_score, f1_score

print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision :", precision_score(y_test, y_pred, average='macro'))
print("Recall :", recall_score(y_test, y_pred, average='macro'))
print("F1 Score :", f1_score(y_test, y_pred, average='macro'))  
```
